```python
books1.set_index(['Userid','Title','Rating'],append=True)
```

```python
books3=books1.reset_index().pivot_table(index='Userid',
                                       columns='Title',
                                       values='Rating')
books3
```

```python
books3.fillna(0,inplace=True)
```

```python
books3
```

```python
#calculating cosine similarities between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user=1-pairwise_distances(books3.values,metric='cosine')
```

```python
user
```

```python
#storing result in a dataframe
user_df=pd.DataFrame(user)
```
